# Cleaning dataset

## Year


```python
# create backup of dataset 
backup=df.copy()
```

```python
df=df[df["year"].str.isnumeric()] # year mai se non numeric values delete 
```

```python
df["year"]=df["year"].astype(int) #values is converted to integer
```

## price

```python
df=df[df["Price"]!="Ask For Price"] #Only numeric values
```

```python
df["Price"]=df["Price"].str.replace(",","")
```

```python
df["Price"]=df["Price"].astype(int)
```

## Kms_driven

```python
df["kms_driven"]
```

```python
#type change
df["kms_driven"]=df["kms_driven"].str.split(" ").str.get(0).str.replace(",","")
```

```python
df=df[df["kms_driven"].str.isnumeric()]
```

```python
df["kms_driven"]=df["kms_driven"].astype(int)
```

```python
df.info()
```

## Fuel type

```python
df=df[~df["fuel_type"].isna()]
```

```python
df
```

## Name

```python
df["name"]=df["name"].str.split().str.slice(0,3).str.join(" ")
```

```python
df=df.reset_index()
```

```python
df.info()
```

```python
df.describe()
```

## Outliers Detection

```python
df["Price"].mean()
```

```python
px.box(data_frame=df,x="Price")
```

```python
df["Price"].skew()
```

```python
Q3=df["Price"].quantile(0.75)
Q1=df["Price"].quantile(0.25)
print(Q3,Q1)
```

```python
IQR=Q3-Q1
print("Iqr is : ",IQR)
```

```python
upper_limit=Q3+(1.5*IQR)
lower_limit=Q1-(1.5*IQR)
print("upper_limit : ",upper_limit)
print("lower_limit : ",lower_limit) #no left side outliers
```

```python
df=df[df["Price"]<upper_limit]
```

```python
px.box(x=df["Price"])
```

```python
df
```

## Model

```python
x=df.drop(columns=["Price"])
y=df["Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(x[["name","company","fuel_type"]])
```

```python
ohe.categories_
```

```python
col_trf=ColumnTransformer(transformers=[("trf1",OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"])],remainder="passthrough")
```

```python
col_trf
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(col_trf,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
print(r2_score(y_pred,y_test))
```
